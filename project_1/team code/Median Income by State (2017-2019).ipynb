```python
#MEDIAN HOUSEHOLD INCOME BY STATE
#Import the libraries we need
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Import the external data excel file in
income = pd.read_excel('./data/ext_med_house_income_state_17_19.xlsx')
```

```python
#Display what our file looks like
income.head(10)
```

```python
#Drop the first 6 unnecessary rows
income.drop(index=income.index[0:6], axis=0, inplace=True)
```

```python
#Keep the figures in current dollars and remove that of 2020 dollars
income.drop(index=income.index[54:], axis=0, inplace=True)
```

```python
#See what it looks like now
income
```

```python
income.columns
```

```python
#Remove unneccesary column name
income.columns = income.iloc[0]
```

```python
income.head()
```

```python
#Change the column header into strings
income.columns = income.columns.astype(str)
```

```python
#Strip out the white spaces
income.columns = income.columns.str.strip()
```

```python
income.reset_index(drop=True)
```

```python
income = income.rename( columns = {
    '2017': 'income_2017',
    '2018': 'income_2018',
    '2019': 'income_2019',
    })
```

```python
income.head()
```

```python
#Isolate the figures for 2017
income_2017 = income[["State","income_2017"]].reset_index()
income_2017
```

```python
#Remove the index column
income_2017 = pd.DataFrame.drop(income_2017, 'index', axis=1)
```

```python
#MEDIAN INCOME DATE FOR HOUSEHOLDS IN 2017
income_2017 = income_2017[2:]
income_2017
```

```python
#Isolate the figures for 2018
income_2018 = income[["income_2018"]].reset_index()
income_2018 = pd.DataFrame.drop(income_2018, 'index', axis=1)
income_2018 = income_2018[2:]
#MEDIAN INCOME DATE FOR HOUSEHOLDS IN 2018
income_2018
```

```python
#Isolate the figures for 2019
income_2019 = income[["income_2019"]].reset_index()
income_2019 = pd.DataFrame.drop(income_2019, 'index', axis=1)
income_2019 = income_2019[2:]
#MEDIAN INCOME DATE FOR HOUSEHOLDS IN 2019
income_2019
```

```python
med_income_17_19 = pd.concat([income_2017, income_2018, income_2019], axis=1)
med_income_17_19
```
