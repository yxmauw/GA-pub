```python
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

### But first: Let's write a function that scales our data for us.
Why do we need to scale our data for clustering algorithms? (thread)
##### Answer: 
Because clustering algorithms being based on distance, are sensitive to magnitude of features. Don't want magnitude of features to affect clustering models performance.

```python
# Write your function to scale data here.
def scale_matrix(X):
    # instantiate standardscaler
    ss = StandardScaler()
    return ss.fit_transform(X) # no need train-test-split for clustering algorithms
```

```python
# Write a function that will plot the results of the three
# clustering algorithms for comparison.

# Your function should take TWO INPUTS:
# 1 - The data
# 2 - The clusterer

# Your function should do TWO THINGS:
# 1 - Print the SILHOUETTE SCORE for your clusters
# 2 - Plot the clusters with appropriate coloring

def plot_results(df, cl): # cl will include parameters for adjustment at function call
    # Scale data
    df_scaled = scale_matrix(df[['x','y']]) # to exclude label column
    # Fit clusterer
    cl.fit(df_scaled)
    # Print Silhouette Score
    print(f'Silhouette score for {cl} model is {silhouette_score(df_scaled, cl.labels_)}')
    # Plot with colors based on model clustering
    # using .labels_ attribute
    df['cluster'] = cl.labels_
    # reference (https://stackoverflow.com/questions/54137694/how-to-print-dataframe-name-in-title-of-a-plot)
    df_name =[x for x in globals() if globals()[x] is df][0] # to isolate name of df for title
    # otherwise it prints dataframe in title
    plt.figure(figsize=(4,4))
    plt.scatter(x=df['x'], y=df['y'], c=df['cluster'], cmap='tab20', s=15)
    plt.title(f'Scatterplot of {df_name} data, color by cluster')
    plt.show()
```

### 5a) Fix bad variable names
We're only going to look at three variables here, and two of them are spelled wrong! Change the lat/long column names to either be spelled correctly or abbreviated as `lat`/`long`. While we're add it, let's change `Price` --> `price`.

```python
mel_renamed = mel.rename(columns={'Lattitude':'lat',
                                  'Longtitude':'long',
                                  'Price':'price'})
```

```python
mel_renamed.columns
```

### 5b) Drop missings
Drop all rows where either `lat`, `long`, or `price` is missing.

```python
mel_renamed.dropna(subset=['lat','long','price'], inplace=True)
```

```python
mel_renamed[['lat','long','price']].isnull().sum() # check no more null values in these features
```

```python
mel_renamed.shape
```

### 5c) Plot lat vs long
What kind of clustering do you want to do?

```python
plt.figure(figsize=(6,6))
plt.scatter(x=mel_renamed['long'], y=mel_renamed['lat'], c=mel_renamed['price'], cmap='tab20', s=2)
plt.title(f'Scatterplot of melbourne data, colored based on price');
```

### 5d) Prepare X matrix by scaling
Why do we need to scale? (thread)
##### Answer: 
Because clustering algorithms being based on distance, are sensitive to magnitude of features. Don't want magnitude of features to affect clustering models performance.

```python
# drop columns with missing values
mel_renamed.dropna(axis=1, inplace=True)
```

```python
mel_renamed.shape # previous shape (20993, 21)
```

```python
mel_renamed.info()
```

```python
ss = StandardScaler()
# scale only numeric features
mel_scaled = ss.fit_transform(mel_renamed.select_dtypes(include=['int64','float64'])) 
```

### 5e) Cluster and plot clusters

```python
# instantiate clustering model
db = DBSCAN(eps=0.4, min_samples=10, metric='l2').fit(mel_scaled)
```

```python
mel_renamed['cluster'] = db.labels_ # new column called 'cluster' with model-assigned labels
```

```python
mel_renamed.plot(
                 kind='scatter',
                 x='long',
                 y='lat',
                 c='cluster',
                 colormap='Spectral',
                 s=2,
                 figsize=(8, 8)
                );
```

### 5f) Do prices differ by cluster?

```python
mel_renamed['cluster'].value_counts() # 226 clusters
```

```python
cluster_price = mel_renamed.groupby('cluster').mean()[['price']]
display(cluster_price.head())
```

```python
plt.figure(figsize=(5,5))
sns.scatterplot(data=cluster_price);
```

If there were distinct clusters, I would be expecting dots to be far apart from each other. However in this scatterplot, most of the dots are close to each other. This means that prices do not differ much between clusters that are close to each other on the plot. However, prices differ widely if comparing clusters that are far from each other on the plot.
